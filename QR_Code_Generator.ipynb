```python
# @title
!pip install pyqrcode
!pip install pypng
!pip install Pillow
import pyqrcode
import png
from PIL import Image
```

### Gerar QR Code

```python
# @title
# Link do site
#@markdown **Insira o link do site para gerar o QR Code:**
link = 'https://github.com/leonardosglima'#@param {type:"string"}

print(f'Link informado: {link}')
url = pyqrcode.create(link)

# Gerando QR Code
img = "qr-code.png"
url.png(img, scale=7)

# Abrindo QR Code
im = Image.open(img)

print('QR Code Gerado! Acesse em:')
im
```
